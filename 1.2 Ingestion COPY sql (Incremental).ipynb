```python
%sql
DROP TABLE IF EXISTS test.default.vendite_incrementali;
CREATE TABLE IF NOT EXISTS test.default.vendite_incrementali;

COPY INTO test.default.vendite_incrementali 
FROM '/Volumes/test/default/test_vol/vendite.parquet'
FILEFORMAT = PARQUET
COPY_OPTIONS ('mergeSchema' = 'true');

SELECT count(distinct sale_id) as distinct_sales, count(*) AS tot FROM test.default.vendite_incrementali
```
