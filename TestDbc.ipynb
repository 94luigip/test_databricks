```python
import pandas as pd

# Leggi i 3 file (un csv, un parquet e un json) caricati precedentemente su un volume
prodotti_df = spark.read.option("header", True).csv("/Volumes/test/default/test_vol/prodotti.csv")
vendite_df = spark.read.parquet("/Volumes/test/default/test_vol/vendite.parquet")
utenti_df = spark.read.json("/Volumes/test/default/test_vol/utenti.json")

# Crea la tabella nel catalog usando python
prodotti_df.write \
    .format("delta") \
    .mode("overwrite") \
    .partitionBy("categoria") \
    .saveAsTable("test.default.prodotti")
```

```python
%sql
CREATE OR REPLACE TABLE test.default.utenti
AS SELECT * 
FROM read_files(
  '/Volumes/test/default/test_vol/utenti.json', 
  format => 'json',
  multiLine => true
)
```

```python
%sql
CREATE OR REPLACE TABLE test.default.vendite
AS SELECT * 
FROM read_files('/Volumes/test/default/test_vol/vendite.parquet', format => 'parquet')
```
